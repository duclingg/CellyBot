```python
import csv
import pandas as pd
```

```python
with open("tiktok_followers.txt", 'r') as file:
    lines = file.readlines()
    
lines = [line.strip() for line in lines]
lines = [line for line in lines if line != "Follow back" and line != "Friends"]
lines = [lines[i] for i in range(len(lines)) if i % 2 != 0]

lines
```

```python
tiktok_followers_df = pd.DataFrame({'Usernames': lines})

tiktok_followers_df.head()
```

```python
tiktok_followers_df.to_csv("tiktok_followers.csv")
```
