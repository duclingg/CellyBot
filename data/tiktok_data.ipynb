```python
import pandas as pd
```

```python
with open("followers.txt", "r") as file:
    lines = file.readlines()
```

```python
lines = [line.rstrip() for line in lines]

for i in range(len(lines)):
    if lines[i] == '':
        lines[i] = None
        
lines
```

```python
df = pd.DataFrame({'username': lines})
df.dropna(inplace=True)
df.head()
```
